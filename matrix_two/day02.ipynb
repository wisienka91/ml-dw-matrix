```
def group_and_barplot(feat_groupby, feat_agg='price_value', agg_funcs=[np.mean, np.median, np.size], feat_sort='mean', top=50, subplots=True):
    return (
        df
            .groupby(feat_groupby)[feat_agg]
            .agg([np.mean, np.median, np.size])
            .sort_values(by=feat_sort, ascending=False)
            .head(top)
    ).plot(kind='bar', figsize=(25, 5), subplots=subplots)
```

```
group_and_barplot('param_marka-pojazdu', 'price_value', feat_sort='size');
```

```
group_and_barplot('param_country-of-origin', 'price_value', feat_sort='size');
```

```
group_and_barplot('param_color', 'price_value', feat_sort='size');
```

```
group_and_barplot('param_kraj-pochodzenia', 'price_value', feat_sort='size');
```

```
group_and_barplot('param_kolor', 'price_value', feat_sort='mean');
```
